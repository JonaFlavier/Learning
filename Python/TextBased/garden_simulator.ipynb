```python
class Plant:
    # attributes: name (str), yield (int), stages of growth ( str arr), current stage (str), ready to harvest(bool)

    def __init__(self, n, y):
        self.name = n
        self.yld = y
        self.growth_stages = ['seed', 'sprout', 'plant','flower', 'harvest ready']
        self.current_growth_stage = self.growth_stages[0]
        self.harvestable = False
        print("Planted {} {}(s)".format(self.yld, self.name))

    def getCurrentGrowthStage(self):
        return self.current_growth_stage

    def grow(self):
        # increment growth stages
        if((self.growth_stages.index(self.current_growth_stage)+1)< len(self.growth_stages)):
            # print("growing from {}".format(self.current_growth_stage))
            tmp_stage = self.current_growth_stage
            self.current_growth_stage = self.growth_stages[self.growth_stages.index(self.current_growth_stage)+1]
            print("{} {}(s) growing from {} to {}".format(self.yld,self.name,tmp_stage, self.current_growth_stage))

            if(self.current_growth_stage == 'harvest ready'):
                self.harvestable = True
                print("{} is ready to harvest".format(self.name))
        else:
            print("{} {}(s) waiting to be harvested".format(self.yld, self.name))

        # if plant becomes harvest ready: set harvestable to True

    def harvest(self):
        if self.harvestable:
            self.harvestable = False
            print("{} {}(s) were harvested".format(self.yld, self.name))
        else:
            print("{}(s) not yet fully developed".format(self.name))


class Carrot(Plant):

    def __init__(self, y):
        super().__init__("Carrot",y)

class Aubergine(Plant):

    def __init__(self, y):
        super().__init__("Aubergine",y)

class Cabbage(Plant):

    def __init__(self, y):
        super().__init__("Cabbage",y)

########################################################################
########################################################################
class Gardener:
    # player that can plant, tend and harvest
    plant_dict = {'Carrot': Carrot,'Aubergine':Aubergine,'Cabbage':Cabbage}

    def __init__(self, name):
        self.name = name
        # list of plants currently planted
        self.planted_plants = []
        # dict that stores the gardener's collection of seeds and harvested plants
        self.inventory = {
            "Seeds":{},
            "Harvested":{}
        }
        print("Update: established role for {} as a Gardener".format(self.name))

    def plant(self):
        # places seed into soil
        # choose specific plant from inventory
        if (len(self.inventory)> 0):
            item = select_item(self.inventory["Seeds"].keys())
            print("{} seeds available in inventory".format(item))
            # reduce its count
            if( self.inventory["Seeds"][item]-1 > 0):
                self.inventory["Seeds"][item] -=1
                print("{} {} seeds left in inventory".format(self.inventory["Seeds"][item], item))

            else:
                # delete key from inventory
                self.inventory["Seeds"].pop(item)
                print("There are no more {}s in inventory".format(item))

            # add to list of growing plants
            self.planted_plants.append(self.plant_dict[item](1))
            print("Planted {} successfully".format(item))
        else:
            print("Inventory is empty")


    def tend(self):
        # ongoing care to plants
        # progress through growth stages
        # loop through planted_plants
        print("{}\nTending to garden...{}".format(div(),div()))
        for plant in self.planted_plants:
            # print("{} {}(s): {} ".format(plant.yld, plant.name, plant.current_growth_stage))
            plant.grow()
        # check if it is harvest ready
        #if not call grow method on plant instance to advance growth stage
        # provide user feedback

    def harvest(self):
        # if plant is harvestable
        # remove from list of growing plants
        # add plant to inventory
        # pick a plant to harvest from planted_plants list
        planted_plants_list = list(map(lambda p: p.name, self.planted_plants) )
        plant = select_item(planted_plants_list)
        # check if plant is harvestable
        plant_idx = planted_plants_list.index(plant)
        plant_obj = self.planted_plants[plant_idx]
        if(plant_obj.harvestable):
            print("Harvesting {} {}(s)...".format(plant_obj.yld, plant_obj.name))
            # pop off planted_list
            harvested_plant = self.planted_plants[plant_idx]
            # add into inventory
            self.inventory[harvested_plant.name] = self.inventory.get(harvested_plant.name, 0) + harvested_plant.yld
            print("{}\nUpdating inventory...{}".format(div(), div()))
            for i in list(self.inventory.keys()):
                print("{} {}(s)".format(self.inventory[i], i))

    def forage_for_seeds(self):
        # select random seed from plant list
        seed = random.choice(list(self.plant_dict.keys()))
        # add picked plant to gardener's inventory
        self.inventory["Seeds"] = self.inventory["Seeds"].get(seed, 0) + 1
        print("Foraged {} added to inventory".format())
        # increment by 1 if it already exists
        # provide feedback to player

########################################################################
#
########################################################################
def div():
    return "\n###############################################"

def select_item(items): # items is a dict or list
    # print("Opening Inventory")
    # helper function to pick an item from inventory
    # display all items in their inventory
    # prompt player to select the number of the item they wish to choose
    # select an item number
    items_list = []
    valid = True
    if (type(items) == dict):
        # print("items are in dictionary: {}".format(items.keys()))
        items_list = list(items.keys())
    elif (type(items) == list):
        # print("list found")
        items_list = items
    else:
        print("Invalid option input")
        valid= False

    # while loop
    error_msg = "Invalid input please try again"
    item_name = ""
    item_idx = 0
    while(valid):
        print("{}\nOptions{}".format(div(), div()))
        for item in items_list:
            print("{} {}".format(items_list.index(item)+1, item))
        try:
            reply = int(input("Select an item (ie. 1, 2..): "))
            if(reply-1 < len(items) and reply-1 >= 0):
                item_name = items_list[reply-1]
                print("{} chosen".format(item_name))
                return item_name
            else:
                print(error_msg)
        except:
            print(error_msg)

def instructions():
    print("Pick an action using its number and plant seeds in your inventory")
def exit():
    print("Exiting game...")
########################################################################
########################################################################


# Create list of valid actions player can do
actions_dict = {"Forage":g.forage_for_seeds, "Plant": g.plant, "Tend": g.tend, "Harvest": g.harvest, "Help": instructions, "Exit Game": exit}

print("{}\nWelcome to the Garden Simulation Game\nGrow as many plants as you wish in your garden{}".format(div(),div(),div()))
# set up the player
player = input("Enter your name: ")
g = Gardener(player)
while(True):
    print("Actions")
    action = select_item(actions_dict)
    actions_list[action]
    if(action == "Exit Game"):
        break




```
